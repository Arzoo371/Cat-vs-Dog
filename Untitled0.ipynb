```python
import zipfile

zip_ref = zipfile.ZipFile('/content/dogs-vs-cats.zip')
zip_ref.extractall('/content')
zip_ref.close()


```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D,Flatten,BatchNormalization
```

```python
# Generators - create batches
train_ds = keras.utils.image_dataset_from_directory(
    directory='/content/train',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256, 256)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory='/content/test',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256, 256)
)
```

```python
# Normalize
def process(image, label):
    image = tf.cast(image / 255., tf.float32)  # Properly divide image by 255 and cast to float32
    return image, label

train_ds = train_ds.map(process)  # Map the process function to the training dataset
validation_ds = validation_ds.map(process)  # Map the process function to the validation dataset
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Flatten, Dense, Dropout

# Creating CNN model
model = Sequential()

# First convolutional layer
model.add(Conv2D(32, kernel_size=(3, 3), padding="valid", activation="relu", input_shape=(256, 256, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'))

# Second convolutional layer
model.add(Conv2D(64, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'))

# Third convolutional layer
model.add(Conv2D(128, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'))

# Flatten the output
model.add(Flatten())

# Fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.1))

# Output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(train_ds, epochs = 10, validation_data = validation_ds)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], color='red', label='train')  # Corrected label syntax
plt.plot(history.history['val_accuracy'], color='blue', label='validation')  # Corrected bracket and key syntax
plt.legend()
plt.show()
```

```python

plt.plot(history.history['loss'], color='red', label='train')  # Corrected label syntax
plt.plot(history.history['val_loss'], color='blue', label='validation')  # Corrected bracket and key syntax
plt.legend()
plt.show()
```
